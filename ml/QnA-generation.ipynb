````python
def generate_qna_pairs(text):
    
    prompt = f'''
    You are a knowledgeable assistant. I will provide you with a piece of text. 
    Your task is to generate 20 unique question-and-answer pairs based on the information contained within the text. 
    Make sure the questions cover a range of details from the text, including key facts, concepts, processes, and any other significant information. 
    Each of the answer to be generated for each question should contain atleast 50 words which are decriptive and answers the question faithfully. Each of the answer and question thus generated should be from the the text provided only.
    
    ormat the output as a JSON array where each element contains a "QUESTION" and an "ANSWER" field.

    Here is the structure of the JSON object for each Q&A pair:
    {{
    "QUESTION": {{
        "question": "<Generated question>",
        "paragraph": "<Relevant paragraph used to generate the question>"
    }},
    "ANSWER": "<Answer>"
    }}

    Text to generate Q&A pairs from: {text}

    The output should be a JSON array with 20 elements, each containing a "QUESTION" object and an "ANSWER".
    '''

    response = model.generate_content(prompt)
    return response
````

````python
import fitz  # PyMuPDF
from PIL import Image

def pdf_to_images(pdf_path, output_folder):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)

    # Iterate through each page in the PDF
    for page_number in range(len(pdf_document)):
        # Get the page
        page = pdf_document.load_page(page_number)

        # Render the page as a pixmap
        pix = page.get_pixmap()

        # Convert the pixmap to Image
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

        # Construct the output image file path (adjust as needed)
        image_path = f"{output_folder}/page_{page_number + 1}.png"

        # Save the image
        img.save(image_path, "PNG")

        print(f"Page {page_number + 1} saved as: {image_path}")

    # Close the PDF document
    pdf_document.close()

# Example usage:
pdf_file = "data/Irrigation Management.pdf"
output_folder = "data/images/Irrigation"

pdf_to_images(pdf_file, output_folder)
````

````python
import base64
# Function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
````

````python
import requests
def gpt4_prompt_with_image(base64_image):
    
    
    prompt = '''
    
    You are a knowledgeable assistant great at framing questions and answers pairs from a given material so as to cover the whole material provided. I will provide you with an image which contains the text to be used. 
    Your task is to generate 15 unique question-and-answer pairs based on the information contained within the text in the image. 
    Make sure the questions cover a range of details from the text, including key facts, concepts, processes, and any other significant information.
    
    Here's the structure of the QnA pairs. Strictly adhere to this structre for QnA pairs.
    
    1) About 33 percent of the generated QnA pairs should contain one word answer types and the question shoud be generated in a sense to match this outcome. Word limit- 10 words
        For example-
        Question- What is the order of the Indian sandalwood tree?
        Answer- The Indian sandalwood tree belongs to the order Santalales.
         
    2)  Another 33 percent of the generated QnA pairs should contain short descriptive answer types and the question shoud be generated in a sense to match this outcome. Word limit- 20-30 words
        For example-
        Question- What is the spiritual and cultural significance of Indian sandalwood?
        Answer- Indian sandalwood holds religious and spiritual significance in India, Sri Lanka, Nepal, and Bhutan.
        
    3)  Last 33 percent of the generated QnA pairs should contain one word answer types and the question shoud be generated in a  sense to match this outcome. Word limit- 30-50 words
        For example-
        Question- What are some medicinal uses of sandalwood oil?
        Answer- Sandalwood oil is used to treat fever, anxiety, bronchitis, diarrhea, fatigue, gallbladder and liver problems, high blood pressure, indigestion, insomnia, low libido, sore throat, and urinary tract infections.
        
    
    Now all 15 of them should contain descriptive questions and answers such that it covers the entirety of the material provided and be unique.
    Remember each generated QnA pair should be completely unique and shouldn't have any coherence with any of the other generated QnA pairs. 

    IMPORTANT: Do generation QnA pairs in which the questions asks to fetch relavant paragraphs since the end user can't see the paragraphs, thus can't be included and accepted as a valid QnA pair.
    
    Here is the structure of the JSON object for each Q&A pair:
    {{
    "QUESTION": {{
        "question": "<Generated question>",
        "paragraph": "<Relevant paragraph used to generate the question>"
    }},
    "ANSWER": "<Answer>"
    }}

    Only print a perfect and a valid JSON array with the above stated number of QnA pairs, each containing a "QUESTION" object and an "ANSWER" and nothing else.
    '''
    
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {openai_api_key}"
    }
    
    payload = {
    "model": "gpt-4o",
    "messages": [
        {
        "role": "user",
        "content": [
            {
            "type": "text",
            "text": prompt
            },
            {
            "type": "image_url",
            "image_url": {
                "url": f"data:image/jpeg;base64,{base64_image}"
            }
            }
        ]
        }
    ],
    # "max_tokens": 300
    }

    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

    # print(response.json())
    res = response.json()['choices'][0]['message']['content']
    return res
````

````python
import requests
def gpt4_prompt_with_text(chunk):
    
    
    prompt = f'''
    
    You are a knowledgeable assistant great at framing questions and answers pairs from the given text so as to cover the whole material provided.
    Your task is to generate strictly 20 unique question-and-answer pairs based on the information contained within the text.
    Make sure the questions cover a range of details from the text, including key facts, concepts, processes, and any other significant information.
    
    Here's the structure of the QnA pairs. Strictly adhere to this structre for QnA pairs.
    
    1) First 10 of the  generated QnA pairs should contain short and descriptive answers around 50 words strictly and should not cross the given limit of words. 
    The questions should be framed that inculcates the essence of the text and is enough to ben answered faithfully within 30-50 words answer.
       
         
    1) Last 10 of the  generated QnA pairs should contain long descriptive answers within the 80-100 words strictly and should not cross the given limit of words. 
    The questions should be framed that inculcates the essence of the text and should be similar to HOTS (Higher Order Thinking Questions) type and is enough to ben answered faithfully within 30-50 words answer.
       
    
    Now all 20 of them should contain descriptive questions and answers such that it covers the entirety of the material provided and be unique.
    Remember each generated QnA pair should be completely unique and shouldn't have any coherence with any of the other generated QnA pairs. 

    IMPORTANT: Do generation QnA pairs in which the questions asks to fetch relavant paragraphs since the end user can't see the paragraphs, thus can't be included and accepted as a valid QnA pair.
    
    Here is the structure of the JSON object for each Q&A pair:
    {{
    "QUESTION": {{
        "question": "<Generated question>",
        "paragraph": "<Relevant paragraph used to generate the question>"
    }},
    "ANSWER": "<Answer>"
    }}
    
    Text provided:
    {chunk}

    Only print a perfectly valid structured JSON array throughout and with the above stated number of QnA pairs, each containing a "QUESTION" object and an "ANSWER" and nothing else.
    '''
    
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {openai_api_key}"
    }
    
    payload = {
    "model": "gpt-4o",
    "messages": [
        {
        "role": "user",
        "content": [
            {
            "type": "text",
            "text": prompt
            },
        ]
        }
    ],
    # "max_tokens": 300
    }


    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

    # print(response.json())
    res = response.json()['choices'][0]['message']['content']
    return res
````
